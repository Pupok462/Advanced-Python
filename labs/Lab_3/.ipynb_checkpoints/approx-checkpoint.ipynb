```python
data1 = pd.read_csv("/home/alex/labs/1.1/B.csv")
data1
```

```python
x = data1["a"]
y = data1["l"]
z = data1["la"]
```

```python
N = 25
zx=0
zy=0
zc=0
x2=0
xy=0
xc=0
yc=0
y2=0
delta=0

for i in range(0,25):
    zc=zc+z[i]
    xc=xc+x[i]
    yc=yc+y[i]
    zx=zx+z[i]*x[i]
    zy=zy+z[i]*y[i]
    x2=x2+x[i]*x[i]
    xy=xy+y[i]*x[i]
    y2=y2+y[i]*y[i]
    
zc=zc/N
xc=xc/N
yc=yc/N
zx=zx/N
zy=zy/N
x2=x2/N
xy=xy/N
y2=y2/N
A=np.array([[x2, -xy, xc], [xy, -y2, yc], [xc, -yc, 1]])
B=np.array([zx, zy, zc])
k=np.linalg.solve(A,B)
print(k)
A=k[0]
C=k[1]
D=k[2]
B=D-C*A
k[2]=C
k[1]=B
print(k)

for i in range(0,25):
    delta=delta+(z[i]-(A*x[i]-C*y[i]+D))**2
delta=delta/N
sigmaA=np.sqrt(2*(delta)/x2)
sigmaC=np.sqrt(2*(delta)/y2)
sigmaD=np.sqrt(2*(delta))
sigmaB=np.sqrt(sigmaD**2+A**2*sigmaC**2+C**2*sigmaA**2)
print(sigmaA, sigmaB, sigmaC)
```

```python
x_new = np.linspace(x.min(), x.max(),500)
def f(x, A, B, C):
    return A+B/(x+C)
y_smooth=f(x_new, A, B, C)

plt.figure(figsize=(8, 6))
plt.plot (x_new,y_smooth)
plt.scatter(x, y, s = 70)
plt.title('lambda(theta)')
plt.xlabel('theta')
plt.ylabel('lambda, A')
plt.tight_layout()
```
